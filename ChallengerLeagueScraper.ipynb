```python
import cassiopeia as cass
```

First let's start with a sample size of only the top challenger players - as this eliminates the bias of subjective skill level, i.e. matches are decided mainly on team comp and matchups and not due to bad decision and random error as commonly seen in low elo games.

We will take the top 50 players from each of the 11 regions, thus 550 players in total

```python
players = {}
for region in cass.data.Region:
    challenger = cass.get_challenger_league(queue='RANKED_SOLO_5x5', region=region)
    challenger = sorted(challenger, key=lambda e: e.league_points, reverse=True)
    summoners = [entry.summoner.name for entry in challenger[:50]]
    players[region] = summoners
```

```python
f = cass.get_summoner(name=players[list(players.keys())[0]][0], region=list(players.keys())[0])
w = cass.get_match_history(summoner=f, end_index=1000)
```
